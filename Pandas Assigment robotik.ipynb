Pandas Conditional Selection and Operation Series

```python
#1. Buatlah data diatas ke dalam sebuah series pandas dimana month berfungsi sebagai label index pada series
data = np.array([408, 943, 902, 117, 389, 435, 793, 356, 651, 993, 786, 157])
series = pd.Series(data, index=["Januari", "Februari", "Maret", "April", "Mei", "Juni", "Juli", "Agustus", "September", "Oktober","November", "Desember"])
series
```

```python
#2. Carilah rata-rata dari nilai series tersebut
series.mean()
```

```python
series > series.mean()
```

```python
series[series > series.mean()]
```

```python
#3. Cek apakah ada elemen series yang memiliki nilai > nilai rata-ratanya. Apabila ada, maka ubah nilai elemen tersebut dengan nilai rata-rata series
mean = series.mean()
def ubah(x):
  if x > mean:
    return mean
  else:
    return x
```

```python
series.apply(ubah)
```

```python
#4. Tampilkan hasil setelah operasi perhitungan pada series
series = series.apply(ubah)
series
```

Reading Data From Databases

```python
#1. Tampilkan semua data pada kolom employees
import sqlite3
import pandas as pd
conn = sqlite3.connect('chinook.db')
database = pd.read_sql('SELECT * FROM employees;', conn)
database
```

```python
#2. Tampilkan data employees yang memiliki city = Calgary
database = pd.read_sql('SELECT * FROM employees WHERE City = "Calgary" ;', conn)
database
```

```python
#3. Apabila City tidak sama dengan Calgary maka ganti data dengan Toronto
def ubah(x):
  if x != "Calgary":
    return "Toronto"
  else:
    return x
```

```python
database['City'] = database['City'].apply(ubah)
database
```

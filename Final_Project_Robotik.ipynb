```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
final_project = pd.read_csv('dataset_final_project.csv')
```

```python
final_project
```

```python
#NO 1 Bagian A
final_project.dtypes
```

```python
final_project['Date'] = pd.to_datetime(final_project['Date'], format="%m/%d/%Y")
```

```python
final_project.dtypes
```

```python
final_project
```

```python
final_project['Time'] = pd.to_datetime(final_project['Time'], format="%H:%M")
```

```python
final_project.dtypes
```

```python
#NO 1 Bagian A
final_project['Branch'] = final_project['Branch'].replace(['A', 'B', 'C'], ['1', '2', '3'])
```

```python
final_project
```

```python
#NO 1 Bagian B
final_project['City'] = final_project['City'].replace(['Yangon', 'Naypyitaw', 'Mandalay'], ['1', '2', '3'])
final_project
```

```python
#NO 1 Bagian C
final_project['Customer type'] = final_project['Customer type'].replace(['Normal', 'Member'], ['1', '2'])
final_project
```

```python
#NO 1 bagian D
final_project['Product line'] = final_project['Product line'].replace(['Electronic accessories', 'Fashion accessories', 'Food and beverages', 'Health and beauty', 'Home and lifestyle', 'Sports and travel'], ['1', '2','3','4','5','6'])
final_project
```

```python
#NO 1 Bagian E
final_project['Payment'] = final_project['Payment'].replace(['Cash', 'Credit card', 'Ewallet'], ['1', '2', '3'])
final_project
```

```python
#No 2 TOTAL
import numpy as np
final_project['Total'] = final_project['Unit price'] * final_project['Quantity']
print(final_project.head())
```

```python
final_project
```

```python
#No 3 TAX
final_project['Tax'] = final_project['Total'] * 5/100
print(final_project.head())
```

```python
final_project
```

```python
#No 2 Final Price
final_project['Final Price'] = final_project['Total'] - final_project['Tax']
print(final_project.head())
```
